```python
def search_top_5(keyword):
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.wait import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    wait=WebDriverWait(driver,500)
    
    #Fetching the search button and clicking on it:
    search=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="x1n2onr6"]//div[@class="_aacl _aacp _aacu _aacx _aada"]')))
    search.click()
    
    #Entering keyword in search box:
    search_box=wait.until(EC.presence_of_element_located((By.XPATH,'//input[@placeholder="Search"]')))
    search_box.send_keys(keyword)
    
    #Extracting list of all suggestions:
    li=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@role="none"]//div[@class="x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1uhb9sk x1plvlek xryxfnj x1iyjqo2 x2lwn1j xeuugli xdt5ytf xqjyukv x1cy8zhl x1oa3qoh x1nhvcw1"]')))
    lis=driver.find_elements(By.XPATH,'//div[@role="none"]//div[@class="x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1uhb9sk x1plvlek xryxfnj x1iyjqo2 x2lwn1j xeuugli xdt5ytf xqjyukv x1cy8zhl x1oa3qoh x1nhvcw1"]')
    names=[]
    for i in range(0,10):
        li=lis[i].text.split()
        names.append(li[0])
    
    
    #making a dictionary of the instagram handles and the number of their followers:
    dic={}
    for i in names:
        string='https://www.instagram.com/'+i+'/'
        driver.get(string)
        followers_element=wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class="_ac2a"]')))
        follower_lis=driver.find_elements(By.XPATH,'//span[@class="_ac2a"]')
        
        num=follower_lis[1].get_attribute('title')
        num=int(num.replace(',',''))
        dic[i]=num
    
    
    
    #sorting the dictionary based on number of followers:
    
    sorted_dic={}
    for w in sorted(dic,key=dic.get,reverse=True):
        sorted_dic[w]=dic[w]

        
    
    return sorted_dic
        
```

```python
def posts_3_days_ago(lis):
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.wait import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    wait=WebDriverWait(driver,500)
    import time
    
    
    post_dic={}
    
    
    for i in lis:
        #visiting every page seperately and checking the posting date of every post:
        string='https://www.instagram.com/'+i
        driver.get(string)
        wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aabd _aa8k  _al3l"]')))
        post_lis=driver.find_elements(By.XPATH,'//div[@class="_aabd _aa8k  _al3l"]')
        count=0
        total_like=0
        for ele in post_lis:
            ele.click()
            #fetching data from the web element having information of posting time:
            time=wait.until(EC.presence_of_element_located((By.XPATH,'//time[@class="_aaqe"]')))
            
            if 'SECOND' in time.text:
                count+=1
            if 'MINUTE' in time.text:
                count+=1
            if 'HOUR' in time.text:
                count+=1
            if 'DAY' in time.text:
                if int(time.text.split()[0])<4:
                    count+=1
                else:
                    break
            
            
            close_button=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="x10l6tqk x160vmok x1eu8d0j x1vjfegm"]')))
            close_button=driver.find_element(By.XPATH,'//div[@class="x10l6tqk x160vmok x1eu8d0j x1vjfegm"]')
            close_button.click()
        
        #adding elements to a dicitionary along with frequency of posts:  
        post_dic[i]=count
        
    return post_dic
        
            
            
```

```python
#plotting bar chart on instagram handle vs No. of posts:
import matplotlib.pyplot as plt

x=[]
y=[]

#plotting instagram handles along X axis and no. of posts along Y axis:
for ele in dic:
    x.append(ele)
    y.append(dic[ele])
plt.bar(x,y,edgecolor="black",width=0.4,color="red",alpha=0.5)
plt.xlabel("Instagram Handle")
plt.ylabel("No.of Post in 3 days")
plt.title("Top Handles vs No.of Posts", color="green")
plt.show()    
```

Q.2 Your friend also needs a list of hashtags that he should use in his posts.
Open the 5 handles you obtained in the last question, and scrape the content of the first 10 posts of each handle.

Prepare a list of all words used in all the scraped posts and calculate the frequency of each word.

Create a csv file with two columns : the word and its frequency

Now, find the hashtags that were most popular among these bloggers

Plot a Pie Chart of the top 5 hashtags obtained and the number of times they were used by these bloggers in the scraped posts.

```python

#Note that this function gives both list of words and average likes in a go, making the code more efficient:


def extract_words_and_likes_posts(lis):
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.wait import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC


    
    wait=WebDriverWait(driver,500)
    
    #Initiating an empty list for list of words and average_like:
    list_of_words=[]
    average_like=[]
    
    
    for i in lis:
        #visiting every page and fetching the likes and list of words:
        string='https://www.instagram.com/'+i
        driver.get(string)
            
        
        #list of posts:
        wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aabd _aa8k  _al3l"]')))
        post_lis=driver.find_elements(By.XPATH,'//div[@class="_aabd _aa8k  _al3l"]')
        post_lis=post_lis[:10]
        
        #initiating ototal_likes count for every page:
        total_like=0
        
        
        #fetching words and likes from every post:
        for ele in post_lis:


            ele.click()
            
            #accessing heading web element to get the text:
            heading=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_a9zs"]')))

            li=heading.text.split('.')
            for i in li:
                sub_li=i.split()
                for j in sub_li:
                    list_of_words.append(j)
            
            #data of likes:
            like=driver.find_element(By.XPATH,'//span[@class="x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj"]')
            lik=like.text
            
            
            
            #refining the data of likes into suitable integer format:
            if len(lik)==1 and'others' not in lik:
                total_like+=0
            if ',' in lik:
                    lik=lik.replace(',','')

            if len(lik.split())>1 and 'others' in lik:
                total_like+=int(lik.split()[-2])+1
            if len(lik.split())>1 and 'others' not in lik:
                total_like+=int(lik.split()[-2])
            close_button=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="x10l6tqk x160vmok x1eu8d0j x1vjfegm"]')))
            close_button=driver.find_element(By.XPATH,'//div[@class="x10l6tqk x160vmok x1eu8d0j x1vjfegm"]')
            close_button.click()
                
        
        

            
        average_like.append(total_like//10)
        
    
    
    #returning the average likes list and list of words:
    return average_like,list_of_words
    
    
```

```python
likes,words=extract_words_and_likes_posts(top_5)
```

#Exporting words data to CSV file:

```python
#exporting data to CSV file:

import pandas as pd
from collections import Counter
c1=dict(Counter(words))

#creating a hashtags dictionary based on frequency:

hashtags={}
for w in sorted(c1,key=c1.get,reverse=True):
    if '#' in w:
        hashtags[w]=c1[w]
_2d_list=[]

#making a dataframe of words and frequency:
for i in c1:
    _2d_list.append([i,c1[i]])

dataframe=pd.DataFrame(_2d_list)

#importing to a CSV file:
dataframe.columns=['words','frequency']
dataframe.to_csv('word-frequency_trial.csv')
```

#Top 5 Hashtags plotted on pie chart:

```python
#Plotting pie chart of top 5 hashtags 
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter

#making a sorted dictiionary of hashtags:
c1=dict(Counter(words))
hashtags={}
for w in sorted(c1,key=c1.get,reverse=True):
    if '#' in w:
        hashtags[w]=c1[w]
count=0

#graph plotting steps from hashtag dictionary:
x=[]
y=[]
for i in hashtags:
    if count==5:
        break
    x.append(i)
    y.append(hashtags[i])
    count+=1
plt.pie(y,labels=x,autopct="%.2f%%")
plt.show()
    
```
